```python
def show_file():  
    # Import the required libraries
    from tkinter import ttk, filedialog
    import numpy
    import pandas as pd

    # Create an instance of tkinter frame
    win = Tk()

    # Set the size of the tkinter window
    win.geometry("1000x700")

    # Create an object of Style widget
    style = ttk.Style()
    style.theme_use('clam')
    # Create a Frame
    frame = Frame(win)
    frame.pack(pady=20)

    win.title("congratulations scraping is done successfully !")

    # Define a function for opening the file
    def open_file():
       filename = filedialog.askopenfilename(title="Open a File", filetype=(("xlxs files", ".*xlsx"),
    ("All Files", "*.")))

       if filename:
          try:
            filename = r"{}".format(filename)
            df = pd.read_excel(filename)

          except ValueError:
             label.config(text="File could not be opened")
          except FileNotFoundError:
             label.config(text="File Not Found")

       # Clear all the previous data in tree
       clear_treeview()
        # Add new data in Treeview widget
       tree["column"] = list(df.columns)
       tree["show"] = "headings"

       # For Headings iterate over the columns
       for col in tree["column"]:
            tree.heading(col, text=col)

       # Put Data in Rows
       df_rows = df.to_numpy().tolist()
       for row in df_rows:
        tree.insert("", "end", values=row)

       tree.pack()
    # Clear the Treeview Widget
    def clear_treeview():
       tree.delete(*tree.get_children())

    # Create a Treeview widget
    tree = ttk.Treeview(frame)

    # Add a Menu
    m = Menu(win)
    win.config(menu=m)

    # Add Menu Dropdown
    file_menu = Menu(m, tearoff=False)
    m.add_cascade(label="Menu", menu=file_menu)
    file_menu.add_command(label="Open Spreadsheet", command=open_file)

    # Add a Label widget to display the file content
    label = Label(win, text='')
    label.pack(pady=20)
    win.mainloop()
    return
```

```python
from tkinter import *
from tkinter import ttk

root = Tk()
root.title('Welcome To Web Scrapping')
root.geometry("500x400")


# Label
ttk.Label(root, text = "Select the website you to scrap from it :", background = 'blue', foreground ="white",
        font = ("Times New Roman", 12)).grid(column = 0, row = 5)


# Creating a list of websites
websites = ['Wuzzuf' ,'Linkedin']

# Create a list of pages_numbers
pages = list(range(1, 1000))

# Create a list of job_search
jobs = ['machine learning' ,'data analysis','software testing']

sel_1=tk.StringVar()
sel_2=tk.StringVar()
user_options_web=tk.StringVar()

def pick_pages_jobs(e):
    if my_combo.get() == "Wuzzuf":
        pages_combo.config(value=pages)
        pages_combo.current(0)
      #  user_options_pages = pages_combo.get()

        job_combo.config(value=jobs)
        job_combo.current(0)


    if my_combo.get() == "Linkedin":
        job_combo.config(value=[""])
        job_combo.destroy
        
        pages_combo.config(value=jobs)
        pages_combo.current(0)
       # user_options_job = job_combo.get()

# datatype of menu text
clicked = tk.StringVar()

        
# Create a drop box
my_combo = ttk.Combobox(root, value=websites, textvariable=user_options_web)
my_combo.grid(column = 1, row = 5, padx = 10, pady = 25)
my_combo.current(0)
#my_combo.pack(pady=20)


# bind the combobox
my_combo.bind("<<ComboboxSelected>>", pick_pages_jobs)

# pages Combo box
pages_combo = ttk.Combobox(root, value=[" "],textvariable=sel_1)
pages_combo.grid(column = 1, row = 8, padx = 10, pady = 25)
pages_combo.current(0)
#pages_combo.pack(pady=20)




# Color Combo box
job_combo = ttk.Combobox(root, value=[" "],textvariable=sel_2)
job_combo.grid(column = 1, row = 10, padx = 10, pady = 25)
job_combo.current(0)
#job_combo.pack(pady=20)


# Create button, it will submit user's chooses
button = Button( root , text = "Submit" , command = root.destroy)
button.place(x=250, y=250)


root.mainloop()

if user_options_web.get() =="Wuzzuf":
    # it will return one csv file without description
    scrap_wuzzuf = object_1.Wuzzuf_scrapping(sel_2.get() , sel_1.get() )
    show_file()
    #print(scrap_wuzzuf.head())

  
elif user_options_web.get() =="Linkedin":
    # it will return two csv files one with description and onather one without it 
    scrap_linkedin_WITH,scrap_linkedin_OUT = object_2.LINKEDIN_Scrapping(job_search = sel_1.get() )
    show_file()
    # call class linkedin (job_search = sel_1.get())


```
